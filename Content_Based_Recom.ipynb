```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_excel("/content/upload_colab.xlsx")
df
```

```python
tf = TfidfVectorizer()

tf.fit(df['Category1'])

tf.get_feature_names_out()
```

```python
tfidf_matrix = tf.fit_transform(df['Category1'])
tfidf_matrix.shape
```

```python
tfidf_matrix.todense()
```

```python
pd.DataFrame(
    tfidf_matrix.todense(),
    columns=tf.get_feature_names_out(),
    index=df.Place
).sample(10, axis=0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix)
cosine_sim
```

```python
cosine_sim_df = pd.DataFrame(
    cosine_sim, index=df.Place, columns=df.Place)

cosine_sim_df
```

```python
cosine_sim_df.to_csv()
```
