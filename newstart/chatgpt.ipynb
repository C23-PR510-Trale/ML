```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
import pandas as pd

# Generate sample data
X = pd.read_csv("dataset3.csv", index_col="user_id")
X
```

```python
# Perform K-means clustering
kmeans = KMeans(n_clusters=5)
kmeans.fit(X.values)
```

```python

# Get cluster labels for each data point
labels = kmeans.labels_
labels
```

```python
# Choose a target point
target_point = np.array([0.5, 0.5])

# Find the nearest cluster center to the target point
nearest_cluster_center = kmeans.transform([target_point]).argmin()

# Get the indices of the points in the nearest cluster
nearest_cluster_indices = np.where(labels == nearest_cluster_center)[0]

# Calculate distances between the target point and all points in the nearest cluster
distances = pairwise_distances_argmin_min([target_point], X[nearest_cluster_indices])[1]

# Sort the distances in ascending order
sorted_indices = np.argsort(distances)

# Select the 10 points with the shortest distances
nearest_points_indices = nearest_cluster_indices[sorted_indices[:10]]
nearest_points = X[nearest_points_indices]

print("Nearest Points:")
print(nearest_points)
```
